```python
#import pustaka yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menyelesaikan persamaan diferensial secara analitik
def pertumbuhan_populasi(t, P0, r):
    """
    Menghitung populasi pada waktu t berdasarkan persamaan eksponensial:
    P(t) = P0 * e^(r * t)

    Parameter:
        t : array-like, titik waktu
        P0: float, populasi awal
        r : float, laju pertumbuhan

    Mengembalikan:
        p : array-like, populasi pada setiap titik waktu
    """
    return P0 * np.exp(r * t)

# Parameter contoh
P0 = 1000   # Populasi awal
r = 0.1     # Laju pertumbuhan (10% per satuan waktu)
t = np.linspace(0, 50, 200)  # Titik waktu dari 0 hingga 50

# Menyelesaikan persamaan
P = pertumbuhan_populasi(t, P0, r)

# Plot hasilnya
plt.figure(figsize=(8, 6))
plt.plot(t, P, label=f"P(t) = {P0} * exp({r} * t)", color="green")
plt.title("Pertumbuhan Populasi Seiring Waktu")
plt.xlabel("waktu(t)")
plt.ylabel("Populasi(P)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def perubahan_suhu_kopi(T,Ta,k,t,T0):
    """
     
    Menghitung perubahan suhu kopi pada k berdasarkan persamaan eksponensial:
    T = Ta + e^-kt * T0

    Parameter
    Ta  : float, suhu ruangan
    k   : float, konstanta laju pendinginan
    t   : array-like, waktu
    T0  : float, suhu awal

    Mengembalikan
    T   : float, suhu akhir
    """
    return Ta + np.exp(-kt) * T0

# Parameter contoh
Ta  = 31 #suhu ruangan
k   = 2
t   = 0.1
T0  = 48

# Menyelesaikan persamaan
perubahan_suhu_kopi(T,Ta,k,t,T0)

# Plot hasilnya
plt.figure(figsize=(6,4))
plt.plot(t,T,k, label = f"T = Ta +np.exp(-kt) * T0", color = "dark")
plt.title(Perubahan Suhu Kopi pada Suhu Ruangan)
plt.xlabel("Waktu")
plt.ylabel("Suhu")
plt.legend()
plt.grid()
plt.show()
    
```
